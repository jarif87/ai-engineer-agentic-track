```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
from IPython.display import Markdown, display
```

```python
quantization_config = BitsAndBytesConfig(load_in_4bit=True)
tokenizer = AutoTokenizer.from_pretrained("/kaggle/input/gemma/transformers/1.1-7b-it/1/")
model = AutoModelForCausalLM.from_pretrained(
    "/kaggle/input/gemma/transformers/1.1-7b-it/1/",
    quantization_config=quantization_config,
    device_map="auto"  # GPU
)
```

```python
def gemma(prompt, max_tokens=256, temp=0.7, top_p=0.9, rep_pen=1.1):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=temp,
        top_p=top_p,
        do_sample=True,
        repetition_penalty=rep_pen,
        pad_token_id=tokenizer.eos_token_id
    )
    # Extract only the *new* text after the prompt
    generated = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated.split(prompt)[-1].strip()
```

```python
prompt_area = """Suggest a single industry or business area that could benefit from an Agentic AI solution.
Respond with ONLY the name of the area. No explanation, no punctuation, no extra words."""

business_area = gemma(prompt_area, max_tokens=15, temp=0.7)
print("Business area:", business_area)
```

```python
prompt_pain = f"""In the {business_area} industry, describe one specific, recurring pain-point that is costly, time-consuming, or error-prone.
Keep it to 1–2 short sentences. Start directly with the pain."""

pain_point = gemma(prompt_pain, max_tokens=120, temp=0.7)
print("\nPain-point:")
display(Markdown(pain_point))
```

```python
prompt_solution = f"""Pain-point in {business_area}:

"{pain_point}"

Propose a concrete Agentic AI solution using 1–3 autonomous agents.
Include:
- What each agent does
- Key tools/APIs it uses
- Expected benefit (quantified if possible)

Be concise and actionable."""

solution = gemma(prompt_solution, max_tokens=400, temp=0.2)
print("\nAgentic AI Solution:")
display(Markdown(solution))
```
